```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
name = ['a','b','c']
data = pd.read_csv('QSO.sed',sep='\t',comment='#',names=name)
data = data[:-1]
x = data['a'].to_numpy().astype(float)
y = data['b'].to_numpy()
data
```

```python
hnu =  9.61225e-04
h = 4.15e-15
nu = hnu / h 
print(r'{:.2e}'.format(nu))
```

```python
plt.plot(x,y*h)
plt.xscale('log')
plt.yscale('log')
plt.xlim(1e-3,1e5)
plt.ylim(1e-17,3e-14)
```
